```python
# Importing Libraries
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt

# Loading Data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

# Data Cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

## 🟩 Count Job Postings by Title and Country (2.4.1)

```python
df_job_title_country = df.pivot_table(
    index='job_title_short',
    columns='job_country',
    aggfunc='size', #counting
    fill_value=0 #replace NaN with 0
)
df_job_title_country
```

---

## 🟨 Salary Stats by Job Title (2.4.2)

```python
salary_stats = df.pivot_table(
    index='job_title_short',
    values='salary_year_avg',
    aggfunc=['mean', 'median', 'min', 'max', 'count']
)

salary_stats
```

---

## 🟥 Sum Salary by Title and Country (2.4.3)

```python
salary_sum = df.pivot_table(
    index='job_title_short',
    columns='job_country',
    values='salary_year_avg',
    aggfunc='sum',
    fill_value=0
)

salary_sum
```
