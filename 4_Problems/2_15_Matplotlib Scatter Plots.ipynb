```python
# Importing Libraries
import ast
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt

# Loading Data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

# Data Cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
df['job_skills'] = df['job_skills'].apply(lambda x: ast.literal_eval(x) if pd.notna(x) else x)
```

## 🟨 Median Yearly Salary vs. Job Count (2.15.1)

```python
job_stats = df.groupby('job_title_short')['salary_year_avg'].agg(
    median_salary='median',
    job_count='count'
)

plt.scatter(job_stats['job_count'], job_stats['median_salary'])
```

---

## 🟨 Top 7 Countries by Median Yearly Salary vs Job Count  (2.15.2)

```python
job_stats_country = df.groupby('job_country')['salary_year_avg'].agg(
    median_salary='median',
    job_count='count'
)

job_stats_country.sort_values(by='median_salary', ascending=False, inplace=True)

top_countries = job_stats_country.head(7)

plt.scatter(top_countries['job_count'], top_countries['median_salary'])

for i in top_countries.index:
    plt.annotate(i,
                 (top_countries.loc[i, 'job_count'],
                  top_countries.loc[i, 'median_salary']))

plt.show()
plt.tight_layout()
```

---

## 🟥 Average Yearly Salary vs. Job Count for Top Roles by Schedule Type (2.15.3)

```python
target_job_titles = ['Data Analyst', 'Data Scientist', 'Data Engineer']


df_filtered = df[
    df['job_title_short'].isin(target_job_titles)
].dropna(subset=['salary_year_avg', 'job_schedule_type'])

job_stats = df_filtered.groupby(
    ['job_title_short', 'job_schedule_type']
)['salary_year_avg'].agg(
    avg_salary='mean',
    job_count='count'
).reset_index()

plt.figure(figsize=(10,6))

for title in target_job_titles:
    subset = job_stats[job_stats['job_title_short'] == title]
    plt.scatter(subset['job_count'], subset['avg_salary'], label=title)

plt.xlabel('Job Count')
plt.ylabel('Average Salary')
plt.title('Average Salary vs Job Count by Schedule Type')

plt.legend()
plt.show()
```
