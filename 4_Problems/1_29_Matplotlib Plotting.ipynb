```python
!pip install -U datasets
import pandas as pd
from datasets import load_dataset

dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

## 🟩 Job Country Bar Chart (1.29.1)

```python
import matplotlib.pyplot as plt

job_counts = df['job_country'].value_counts().head(10)
plt.barh(job_counts.index, job_counts.values)
plt.show()
```

---

## 🟨 Average Salary Horizontal Bar Chart (1.29.2)

```python
avg_salary = df.groupby('job_title_short')['salary_year_avg'].mean().sort_values()
plt.barh(avg_salary.index, avg_salary.values)
plt.show()
```

---

## 🟥 Average Salary Trend Line Plot (1.29.3)

```python
df['job_posted_month'] = df['job_posted_date'].dt.month
avg_salary_trend = df.groupby('job_posted_month')['salary_year_avg'].mean()
plt.plot(avg_salary_trend.index, avg_salary_trend.values)
plt.show()
```
